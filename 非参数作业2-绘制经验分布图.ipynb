# 1.找数据，画出经验分布

```r
data(iris)
df = iris[,1]  #导入数据:Sepal.Length;length(df)=150
df.sort = sort(df)
df.rank = rank(df.sort) 
df.cdf = df.rank/length(df)
```

```r
plot(df.sort,df.cdf)
N = length(df)
segments(df.sort[1:(N-1)],df.cdf[1:(N-1)] , df.sort[2:N],df.cdf[1:(N-1)])
alpha = 0.05
band = sqrt(1/(2*length(df))*log(2/alpha))
lower.95 = df.cdf-band
upper.95 = df.cdf+band
lines(df.sort,lower.95,lty=2)
lines(df.sort,upper.95,lty=2)
```

# 1.10

```r
x=rexp(100,1) # 生成 100 个均值为 1 的指数分布随机数 
hist(x,prob=T,col=gray(0.9),main="exp(1/10)") #gray表示灰色里面参数表示灰度
curve(dexp(x,1/10),add=T,col="red") #添加指数分布密度函数线(模拟指数分布曲线图)
```

```r
df = x
df.sort = sort(df)
df.rank = rank(df.sort) 
df.cdf = df.rank/length(df)
df.cdf = log(1-df.cdf)
sd(df.cdf[-(length(df.cdf))]) # 标准差
```

```r
plot(df.sort,df.cdf)
N = length(df)
segments(df.sort[1:(N-1)],df.cdf[1:(N-1)] , df.sort[2:N],df.cdf[1:(N-1)])
alpha = 0.05
band = sqrt(1/(2*length(df))*log(2/alpha))
lower.95 = df.cdf-band
upper.95 = df.cdf+band
lines(df.sort,lower.95,lty=2)
lines(df.sort,upper.95,lty=2)
```
